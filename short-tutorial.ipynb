## 4 - Estruturas de dados: listas

```python
# Criando listas
letras = ["a", "b", "c", "d"]
print(letras)
matriz = [[0, 1], [1, 2]]
print(matriz)
zeros = [0] * 5
print(zeros)
combinando = zeros + letras
print(combinando)
numeros = list(range(20))
print(numeros)
```

```python
# Acessando ítens da lista
letras[0]  # "a"
```

```python
letras[1:3]
```

```python
# sub-selecionando listas 
print(letras[0:3])   # "a", "b", "c"
print(letras[:3])    # "a", "b", "c"
print(letras[0:])    # "a", "b", "c", "d"
print(letras[:])     # "a", "b", "c", "d"
print(letras[::2])   # "a", "c"
print(letras[::-1])  # "d", "c", "b", "a" 
```

```python
# Adicionando letras 
letras.append("f")
print(letras)
letras.insert(0, "-")
print(letras)
```

```python
# Removendo letras 
letras.pop()
print(letras)
letras.pop(0)
print(letras)
letras.remove("b")
print(letras)
```

```python
# encontrando elementos
"f" in letras
```

```python
"a" in letras
```

## 7 - Estruturas de repetição

```python
# para cada ítem em uma lista
for n in range(1, 10): 
    # imprime o valor
    print(n)    
```

```python
# enquanto a condição
# não for satisfeita 
# repita as operações
n = 0
while n < 10: 
    print(n)
    # adiciona 1 à variável n
    n += 1
```

## 8 - Funções

```python
# definir uma função para executar
# uma tarefa específica
def minha_soma(valor_1, valor_2):
     print(valor_1, '+', valor_2, '=', valor_1+valor_2)

# executar a função
minha_soma(10, 3)
```

```python
# Uma pode retornar um valor
def minha_soma_nova(valor_1, valor_2):
     return valor_1+valor_2
    
a = minha_soma(10, 3)
b = minha_soma_nova(10, 3)
```

```python
# a função 'minha_soma' executa uma impressão
# e não retorna o valor, por isso
# a variável 'a' está vazia
a
```

```python
# a função 'minha_soma_nova' executa uma impressão
# e retorna o valor, por isso
# a variável 'b' não está vazia
b
```

## 9 - Entrada e saída de arquivos (I/O)

```python
f = open("arquivo_1.csv", "w") # a função open cria um manipulador de arquivos (handler),
                               # que é atribuído para a variável f, 
                               # enquanto "w" é o parâmetro que indica o modo (w para write)
for item in range(6):
    f.write(str(item)) # escreve o número para o arquivo
    # adicione uma caractere que faz com que o arquivo mude de linha
    f.write("\n") 
    # alternativamente podemos fazer:
    # f.write(str(item)+"\n") # como vimos anteriormente `+` concatena caracteres
f.close() # fecha o manipulador de arquivos por motivos de segurança
```

```python
f = open("arquivo_1.csv", "r") # modo leitura
conteudo = [item for item in f] # lembra se do list comprehension? Se não, volte a seção anterior
print(conteudo)
```

### D - Processando os dados

```python
# método 'describe' apresenta o súmario estatístico
# para colunas selecionadas

# colunas com texto

sy1617[['School_Type', 'Primary_Category']].describe()
```
